<h1>ACTIVATION FUNCTIONS</h1>

1. Sigmoid Function
- Scales the given value between 0 to 1.
- Generates a S-Shaped Curve

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x_data = np.linspace(-10, 10, 1000)
```

```python
data_sigmoid = [sigmoid(x) for x in x_data]
```

```python
plt.plot(x_data, data_sigmoid)
plt.xlabel("X")
plt.ylabel("Sigmoid(X)")
```

- As can be seen from the graph above, for any value of x the value generated by sigmoid function is between 0 and 1.
- For small values of x the sigmoid value is ≈0
- For large values of x the sigmoid value is ≈1

2. Tanh Function
- Scales the given value between -1 to +1
- Genrates S-Shaped curve just like the Sigmoid Function

```python
def tanh(x):
    numerator = 1 - np.exp(-2 * x)
    denominator = 1 + np.exp(-2 * x)
    return numerator / denominator
```

```python
x_data = np.linspace(-5, 5, 10000)
```

```python
data_tanh = [tanh(x) for x in x_data]
```

```python
plt.plot(x_data, data_tanh)
plt.xlabel("X")
plt.ylabel("Tanh(X)")
```

- As can be seen from the graph above, for any value of x the value generated by tanh function is between 0 and 1.
- For small values of x the sigmoid value is ≈-1
- For large values of x the sigmoid value is ≈+1

3. ReLU(Rectified Linear Unit) Function
- Scales the given value between 0 to infinity
 $$  f(x)=
    \begin{cases}
      0 , & \text{if}\ x<0 \\
      x, & \text{otherwise}
    \end{cases}
    $$
- Also expressed as:
$$ f(x) = max(0, x) $$

```python
def ReLU(x):
    if x < 0:
        return 0
    else:
        return x
```

```python
x_data = np.linspace(-10, 10, 100)
```

```python
data_ReLU = [ReLU(x) for x in x_data]
```

```python
plt.plot(x_data, data_ReLU)
plt.xlabel("X")
plt.ylabel("ReLU(X)")
```

- As can be seen from the graph above, for any value of x the value generated by ReLU function starts from 0 scales upto infinity for sufficiently large value of x.
- For -ve values of x the sigmoid value is 0
- For +ve values of x the sigmoid value is x

3. Leaky ReLU(Rectified Linear Unit) Function
- Same as ReLU for +ve input
- However, instead of converting -ve input to zero like ReLU, it introduces a small slope(α) for negative values
 $$  f(x)=
    \begin{cases}
      \alpha x, & \text{if}\ x<0 \\
      x, & \text{otherwise}
    \end{cases}
    $$
- The value for α is usually set to 0.01
- The value of α can also be set to random, in which case the function is called Randomized ReLU function

```python
# Computation of large array take long time
# So for generating a convincing graph for Leaky ReLU we will use a larger value of alpha
def leaky_ReLU(x, alpha=0.1):
    if x < 0:
        return alpha * x
    else:
        return x
```

```python
x_data = np.linspace(-1000, 1000, 2000)
```

```python
data_leaky_ReLU = [leaky_ReLU(x) for x in x_data]
```

```python
plt.plot(x_data, data_leaky_ReLU)
plt.xlabel("X")
plt.ylabel("Leaky ReLU(X)")
```

4. ELU(Exponential Linear Unit) Function
- Similar to Leaky ReLU i.e has a small slope for -ve values.
- However, insted of a straight line, it has a log curve.
 $$  f(x)=
    \begin{cases}
      \alpha (e^x-1), & \text{if}\ x<0 \\
      x, & \text{otherwise}
    \end{cases}
    $$
- The value for α is usually set to 0.01

```python
# Computation of large array take long time
# So for generating a convincing graph for Leaky ReLU we will use a larger value of alpha
def ELU(x,alpha=1):
    if x < 0:
        return alpha * (np.exp(x) - 1)
    else:
        return x
```

```python
x_data = np.linspace(-10, 10, 20)
```

```python
data_ELU = [ELU(x) for x in x_data]
```

```python
plt.plot(x_data, data_ELU)
plt.xlabel("X")
plt.ylabel("ELU(X)")
```

5. Swish Function
- Unique than other activation functions because it is non-monotnic.
- It is neither always increasing nor always decreasing.
- Found to provide better performance than ReLU.
 $$  f(x)= x \sigma(x) \text{\hspace{5mm} where } \sigma(x) \text{ is a sigmoid function} $$
- Can also be pramaterized as:
 $$  f(x)= 2x \sigma(\beta x) $$
- For β=0,
 $$  f(x)= x $$
- For β=infinity,
$$  f(x)= 2max(0, x) $$

```python
def swish(x, beta):
    return 2 * x * sigmoid(beta * x)
```

```python
x_data = np.linspace(-1000, 1000, 2000)
```

```python
data_swish = [swish(x, 0.01) for x in x_data]
```

```python
plt.plot(x_data, data_swish)
plt.xlabel("X")
plt.ylabel("Swish(X)")
```

5. Softmax Function
- Generalization of the sigmoid function
- Usually applied to the final layer of the network when performing multi-class classification.
- Gives probabilities of each class for being output and thus, the sum of softmax values will always be equal to 1.
 $$ f(x_i) = \frac{e^{x_i}}{\Sigma _j e^{x_j}} $$
 $$ \begin{bmatrix}
 0.5 \\
 1.3 \\
 1.1 \\
 \end{bmatrix}
 \text{ ---[SOFTMAX]---> }
 \begin{bmatrix}
 0.198 \\
 1.440 \\
 1.360 \\
 \end{bmatrix}
 $$

```python
def softmax(x):
    return np.exp(x) / np.exp(x).sum(axis=0)
```

<h1>FORWARD PROPAGARION</h1>

```python
def forward_prop(X, Wxh, Why):
    z1 = np.dot(X, Wxh) + bh
    a1 = sigmoid(z1)
    z2 = np.dot(a1, Why) + by
    y_hat = sigmoid(z2)
    return z1, a1, z2, y_hat
```

<h4>Building a neural network from scratch</h4>

- Here we will build a neural network that learns the XOR logic.<br>
- It will have two nodes in input layer, a hidden layer with five nodes and an ouput layer with one node.<br>

Deifing the XOR gate truth table

```python
X = np.array([ [0, 1], [1, 0], [1, 1],[0, 0] ])
y = np.array([ [1], [1], [0], [0]])
```

Defining the nodes in each layer

```python
num_input = 2
num_hidden = 5
num_output = 1
```

Random weights and bias initialization for hidden layer.

```python
Wxh = np.random.randn(num_input,num_hidden)
bh = np.zeros((1,num_hidden))
```

Random weights and bias initialization for ouput layer.

```python
Why = np.random.randn (num_hidden,num_output)
by = np.zeros((1,num_output))
```

Now we define the derivative of sigmoid function

```python
def sigmoid_derivative(z):
    '''
    Returns the value of 1st derivative of the sigmoid function for a value z
    Used by the backpropogation algorithm
    '''
    return np.exp(-z)/((1+np.exp(-z))**2)
```

Defining the cost function

```python
def cost_function(y, y_hat):
    J = 0.5*sum((y-y_hat)**2)
    return J
```

Now we set the learning rate and the number of iterations

```python
alpha = 0.01
num_iterations = 5000
```

Finally we train our network with the XOR data

```python
cost =[]
for i in range(num_iterations):
    z1,a1,z2,y_hat = forward_prop(X,Wxh,Why)
    dJ_dWxh, dJ_dWhy = backword_prop(y_hat, z1, a1, z2)
    #update weights
    Wxh = Wxh -alpha * dJ_dWxh
    Why = Why -alpha * dJ_dWhy
    #compute cost
    c = cost_function(y, y_hat)
    cost.append(c)
```

```python
plt.grid()
plt.plot(range(num_iterations),cost)
plt.title('Cost Function')
plt.xlabel('Training Iterations')
plt.ylabel('Cost')
```

- Our main motive when training a neural network is to reduce the cost function.
- The above graph clearly shows that as the number of training iterations increases, the cost function keeps of decreasing.
- At around 4500-5000 iterations the depression of the curve seems very low which indicates the peak learning point of the network.
- The constant cost value indicates that the model has stopped learning or is learning very very slowly.
